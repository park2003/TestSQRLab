```julia
import Pkg
using Revise
using SQRLab
```

```julia
t = create_standard_chamber()
```

```julia

using IonSim
using QuantumOptics
using PyPlot
```

```julia

t_cal_max = 50e-6 
tspan = 0:t_cal_max/1000:t_cal_max
# Solve the time evolution using the Schrdinger equation solver from QuantumOptics.jl
tout, excited_pop = run_simulation(t, tspan)


# Create the plot using PyPlot
figure(figsize=(10, 6))
plot(tout.* 1e6, real.(excited_pop), label="Excited State Population", color="C0", linewidth=2)
xlabel("Time (s)")
ylabel("Population")
title("Carrier Rabi Oscillations")
ylim(0, 1.05)
grid(true, linestyle="--")
legend()
display(gcf())
```

```julia
H = hamiltonian(t, rwa_cutoff=Inf)
```

```julia

t_cal_max = 50e-6 
tspan = 0:t_cal_max/1000:t_cal_max
# Solve the time evolution using the Schrdinger equation solver from QuantumOptics.jl
tout, excited_pop = run_simulation_hamiltonian(t, H, tspan)


# Create the plot using PyPlot
figure(figsize=(10, 6))
plot(tout.* 1e6, real.(excited_pop), label="Excited State Population", color="C0", linewidth=2)
xlabel("Time (s)")
ylabel("Population")
title("Carrier Rabi Oscillations")
ylim(0, 1.05)
grid(true, linestyle="--")
legend()
display(gcf())
```
